### Replacing null values with median

Except num-of-doors all others with missing value are of foat or int type. num-of-doors is object type. 

```python
#maping no-of-doors 
data['num-of-doors']=data['num-of-doors'].map({'two':2,'four':4})
```

```python
#maping num-of-cylinders
data['num-of-cylinders']=data['num-of-cylinders'].map({'two':2,'three':3,'four':4,'five':5,'six':6,'eight':8,'twelve':12})
```

```python
data['normalized-losses']=data['normalized-losses'].fillna((data['normalized-losses'].median()))
data['horsepower']=data['horsepower'].fillna((data['horsepower'].median()))
data['peak-rpm']=data['peak-rpm'].fillna((data['peak-rpm'].median()))
data['bore']=data['bore'].fillna((data['bore'].median()))
data['stroke']=data['stroke'].fillna((data['stroke'].median()))
data['price']=data['price'].fillna((data['price'].median()))
data['num-of-doors']=data['num-of-doors'].fillna((data['num-of-doors'].median()))
```

#### Multicollinearity

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
variables=data2[[ 'width','curb-weight',  'engine-size',  'horsepower',  'city-mpg']]
vif=pd.DataFrame()
vif['VIF']=[variance_inflation_factor(variables.values,i) for i in range (variables.shape[1])]
vif['features']=variables.columns
vif
```

### Dropping unwanted columns

```python
data_no_multi=data2.drop(['symboling', 'normalized-losses','num-of-doors','wheel-base', 'length','height','bore','stroke',
       'compression-ratio','peak-rpm','engine-location','aspiration','fuel-type','fuel-system'], axis=1)
data_no_multi.head()
```

### Creating dummy variables

#### Dummy Variable Encoding for body-style column and drive-wheel

```python
data_dummy=pd.get_dummies(data_no_multi,columns=["body-style", "drive-wheels"], prefix=["body", "drive"])
data_dummy.head()
```

#### Custom Binary Encoding for engine-type

```python
data_dummy['engine-type'].value_counts()
```

Most of the values are ohc. So we can take ohc as 1 and the rest as 0.

```python
data_dummy['ohc_code']=np.where(data_dummy['engine-type'].str.contains('ohc'),1,0)

#drop the engine-type variable
data3=data_dummy.drop('engine-type',axis=1)
```

#### Ordinal Encoding for make column

```python
from sklearn.preprocessing import OrdinalEncoder

ord_enc = OrdinalEncoder()
data3["make_code"] = ord_enc.fit_transform(data3[["make"]])

#drop the make variable
data_preprocessed=data3.drop('make',axis=1)
data_preprocessed.head()
```

##### Now we can do multiple linear regression 

```python
target=data_preprocessed['log_price']
inputs=data_preprocessed.drop(['log_price'],axis=1)
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(inputs,target,test_size=.2, random_state=42,shuffle=True)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

### Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
reg.fit(x_train,y_train)
```

```python
reg.score(x_train,y_train)
```

```python
reg.coef_
```

```python
reg.intercept_
```

### Adjusted R-square

```python
r2=reg.score(x_train,y_train)
n=x_train.shape[0]
p=x_train.shape[1]
adjusted_r2=1-(1-r2)*(n-1)/(n-p-1)
adjusted_r2
```

### F-stats and P value

```python
from sklearn.feature_selection import f_regression
f_stats=f_regression(x_train,y_train)[0]
p_value=f_regression(x_train,y_train)[1]
```

### Creating a summary table

```python
reg_summ=pd.DataFrame(data=x_train.columns.values,columns=['Features'])
reg_summ['coefficients']=reg.coef_
reg_summ['p_values']=p_value.round(3)
reg_summ
```

### Prediction

```python
y_hat=reg.predict(x_train)
```

##### Plotting

```python
plt.scatter(y_train,y_hat)
plt.ylabel('prediction')
plt.xlabel('Target')
plt.show()
```

```python
sns.distplot(y_train-y_hat)
plt.title("Residuals pdf")
```

### Testing

```python
y_hat_test=reg.predict(x_test)
plt.scatter(y_test,y_hat_test)
plt.ylabel('prediction')
plt.xlabel('Target')
plt.show()
```
